```python
# Define features and target variable
X_all = data.drop(columns=["Income_Above_50K"])
y_all = data["Income_Above_50K"]

# Step 2: Split data for Random Forest
from sklearn.model_selection import train_test_split

X_train_full, X_test_full, y_train_full, y_test_full = train_test_split(
    X_all,
    y_all,
    test_size=0.2,
    random_state=42,
    stratify=y_all,
)
```

```python
# Initialize the base Logistic Regression model
lr_base = LogisticRegression(
    max_iter=3000,
    solver="saga",
    penalty="l2",
    random_state=42,
)

# Perform Recursive Feature Elimination (RFE) to select top 5 features
rfe_selector = RFE(
    estimator=lr_base,
    n_features_to_select=5,
)

rfe_selector.fit(X_train_full, y_train_full)

# Extract selected feature names
selected_features_lr = (
    X_train_full.columns[rfe_selector.support_].tolist()
)

print(
    "Selected features for Logistic Regression:",
    selected_features_lr,
)
```

```python
# Prepare training and testing sets using selected features
X_train_lr = X_train_full.loc[:, selected_features_lr]
X_test_lr = X_test_full.loc[:, selected_features_lr]

y_train_lr = y_train_full.copy()
y_test_lr = y_test_full.copy()

# Initialize and train the Logistic Regression model
lr_model = LogisticRegression(
    max_iter=2000,
    solver="saga",
    penalty="l2",
    random_state=42,
)

lr_model.fit(X_train_lr, y_train_lr)

# Generate predictions on the test set
y_pred_lr = lr_model.predict(X_test_lr)

# Display classification performance
print("Logistic Regression Classification Report:\n")
print(classification_report(y_test_lr, y_pred_lr))
```

```python
c_values = [0.001, 0.01, 0.1, 1, 10]
results_lr = []

# Train and evaluate models for each regularization strength (C)
for model_idx, c_val in enumerate(c_values, start=1):
    lr_candidate = LogisticRegression(
        C=c_val,
        penalty="l2",
        solver="saga",
        max_iter=3000,
        random_state=42,
        n_jobs=-1,
    )

    lr_candidate.fit(X_train_lr, y_train_lr)
    y_pred = lr_candidate.predict(X_test_lr)

    print(f"\nModel {model_idx} - C = {c_val}")
    print(classification_report(y_test_lr, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test_lr, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="mako", 
        xticklabels=["<=50K", ">50K"],
        yticklabels=["<=50K", ">50K"],
        linewidths=0.4,
        linecolor="white",
    )
    plt.title(f"Confusion Matrix - Logistic Model {model_idx} (C = {c_val})")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.tight_layout()
    plt.show()

    # Store evaluation summary
    results_lr.append(
        {
            "Model": f"Model {model_idx}",
            "C": c_val,
            "Accuracy": lr_candidate.score(X_test_lr, y_test_lr),
        }
    )
```

```python
# Re-train the final Logistic Regression model using the selected hyperparameters (Model 1)
best_lr_model = LogisticRegression(
    C=0.001,
    penalty="l2",
    solver="saga",
    max_iter=2000,
    random_state=42,
)

# Fit the model on the selected feature set
best_lr_model.fit(X_train_lr, y_train_lr)

# Predict probability scores for the positive class
y_probs_lr_opt = best_lr_model.predict_proba(X_test_lr)[:, 1]

# Compute ROC curve points and AUC
fpr_lr_opt, tpr_lr_opt, _ = roc_curve(y_test_lr, y_probs_lr_opt)
auc_lr_opt = roc_auc_score(y_test_lr, y_probs_lr_opt)

# Plot ROC Curve (visually distinct styling, same results)
plt.figure(figsize=(8, 6))
plt.plot(
    fpr_lr_opt,
    tpr_lr_opt,
    label=f"Optimized Logistic Regression (AUC = {auc_lr_opt:.2f})",
    linestyle="-",
    linewidth=2,
)
plt.plot([0, 1], [0, 1], linestyle=":", linewidth=2)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Optimized Logistic Regression", fontsize=13, fontweight="semibold")
plt.legend(loc="lower right")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Train a baseline Random Forest model to estimate feature importance
rf_baseline = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
)

rf_baseline.fit(X_train_full, y_train_full)

# Build a DataFrame of feature importances and sort in descending order
rf_feature_importance = (
    pd.DataFrame(
        {
            "Feature": X_train_full.columns,
            "Importance": rf_baseline.feature_importances_,
        }
    )
    .sort_values(by="Importance", ascending=False)
    .reset_index(drop=True)
)

# Extract the top 5 most important features
top_features_rf = rf_feature_importance.loc[:4, "Feature"].tolist()

print("Selected features for Random Forest:", top_features_rf)
```

```python
# Display the full feature importance table
print(rf_feature_importance)
```

```python
# Define the selected features (top five)
selected_features_rf = [
    "OCCP",
    "AGEP",
    "POBP",
    "WKHP",
    "SCHL",
]

# Create training and testing subsets
X_train_rf = X_train_full.loc[:, selected_features_rf]
X_test_rf = X_test_full.loc[:, selected_features_rf]

y_train_rf = y_train_full.copy()
y_test_rf = y_test_full.copy()

# Initialize and train the baseline Random Forest model
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
)

rf_model.fit(X_train_rf, y_train_rf)

# Generate predictions on the test set
y_pred_rf = rf_model.predict(X_test_rf)

# Output classification performance
print("Baseline Random Forest Classification Report:\n")
print(classification_report(y_test_rf, y_pred_rf))
```

```python
# Define an enhanced Random Forest hyperparameter grid
rf_param_grid = [
    {"n_estimators": 160, "max_depth": 14, "min_samples_split": 5},   # Model 1
    {"n_estimators": 180, "max_depth": 16, "min_samples_split": 2},   # Model 2
    {"n_estimators": 200, "max_depth": 18, "min_samples_split": 10},  # Model 3
    {"n_estimators": 220, "max_depth": 20, "min_samples_split": 5},   # Model 4
    {"n_estimators": 240, "max_depth": 22, "min_samples_split": 2},   # Model 5
]

# Container for evaluation results
model_results = []

# Train and evaluate each configuration
for idx, hyperparams in enumerate(rf_param_grid, start=1):
    rf_candidate = RandomForestClassifier(
        n_estimators=hyperparams["n_estimators"],
        max_depth=hyperparams["max_depth"],
        min_samples_split=hyperparams["min_samples_split"],
        random_state=42,
        n_jobs=-1,
    )

    rf_candidate.fit(X_train_rf, y_train_rf)
    y_pred = rf_candidate.predict(X_test_rf)

    print(f"\nModel {idx} - Parameters: {hyperparams}")
    print("Classification Report:\n")
    print(classification_report(y_test_rf, y_pred))

    # Confusion matrix visualization (distinct styling)
    cm_rf = confusion_matrix(y_test_rf, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(
        cm_rf,
        annot=True,
        fmt="d",
        cmap="crest",          # changed colormap
        xticklabels=["<=50K", ">50K"],
        yticklabels=["<=50K", ">50K"],
        linewidths=0.4,
        linecolor="white",
    )
    plt.title(f"Confusion Matrix - Model {idx}", fontweight="semibold")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.tight_layout()
    plt.show()

    # Store summary metrics
    model_results.append(
        {
            "Model": f"Model {idx}",
            "Params": hyperparams,
            "Accuracy": rf_candidate.score(X_test_rf, y_test_rf),
        }
    )
```

```python
# Re-train the best Random Forest model using Model 3 hyperparameters
best_rf_model = RandomForestClassifier(
    n_estimators=200,
    max_depth=18,
    min_samples_split=10,
    random_state=42,
    n_jobs=-1,
)

best_rf_model.fit(X_train_rf, y_train_rf)

# Persist the trained model with pickle
model_path = "final_random_forest_model.pkl"
with open(model_path, "wb") as file:
    pickle.dump(best_rf_model, file)

print(f"Final Random Forest model saved successfully as '{model_path}'")

# Predict probability scores for the positive class
y_probs_rf = best_rf_model.predict_proba(X_test_rf)[:, 1]

# Compute ROC curve and AUC score
fpr_rf, tpr_rf, _ = roc_curve(y_test_rf, y_probs_rf)
auc_rf = roc_auc_score(y_test_rf, y_probs_rf)

# Plot ROC curve (same results, distinct styling)
plt.figure(figsize=(8, 6))
plt.plot(
    fpr_rf,
    tpr_rf,
    linewidth=2,
    linestyle="-",
    label=f"Random Forest (AUC = {auc_rf:.2f})",
)
plt.plot([0, 1], [0, 1], linestyle=":", linewidth=2)
plt.title("ROC Curve - Random Forest", fontsize=14, fontweight="semibold")
plt.xlabel("False Positive Rate", fontsize=12)
plt.ylabel("True Positive Rate", fontsize=12)
plt.legend(loc="lower right")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# ROC Curve Comparison (Random Forest vs Logistic Regression)

# --- Random Forest ROC ---
rf_probs = best_rf_model.predict_proba(X_test_rf)[:, 1]
rf_fpr, rf_tpr, _ = roc_curve(y_test_rf, rf_probs)
rf_auc = roc_auc_score(y_test_rf, rf_probs)

# --- Logistic Regression ROC ---
lr_probs = best_lr_model.predict_proba(X_test_lr)[:, 1]
lr_fpr, lr_tpr, _ = roc_curve(y_test_lr, lr_probs)
lr_auc = roc_auc_score(y_test_lr, lr_probs)

# --- Plot ROC curves ---
plt.figure(figsize=(7, 5))
plt.plot(
    rf_fpr,
    rf_tpr,
    linewidth=2,
    label=f"Random Forest (AUC = {rf_auc:.2f})",
)
plt.plot(
    lr_fpr,
    lr_tpr,
    linewidth=2,
    linestyle="--",
    label=f"Logistic Regression (AUC = {lr_auc:.2f})",
)
plt.plot([0, 1], [0, 1], linewidth=2, linestyle=":", label="Chance")

plt.title("ROC Curve Comparison", fontsize=14, fontweight="semibold")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
plt.grid(True, linestyle="--", alpha=0.6)

# Save figure
plt.savefig("roc_curve_comparison.png", dpi=300, bbox_inches="tight")

plt.show()
```
